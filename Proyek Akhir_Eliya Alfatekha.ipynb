# Data Preprocessing

```python
full_data = [train, test]
```

```python
for data in full_data:
    obj_job = {'Pekerja Bebas Bukan Perusahan':0,'Wiraswasta':1,
               'Tidak Pernah Bekerja':2,
               'Tanpa di Bayar':3,
               'Pemerintah Provinsi':4,
               'Pemerintah Negara':5,
               'Pemerintah Lokal':6,
               'Pekerja Bebas Perusahaan':7}
    data['Kelas Pekerja'] = data['Kelas Pekerja'].replace(obj_job)
```

```python
for data in full_data:
    data.loc[data['Umur'] <=17, 'Umur'] = 0
    data.loc[(data['Umur'] > 17) & (data['Umur'] <= 34), 'Umur'] = 1
    data.loc[(data['Umur'] > 32) & (data['Umur'] <= 51), 'Umur'] = 2
    data.loc[(data['Umur'] > 51) & (data['Umur'] <= 68), 'Umur'] = 3
    data.loc[(data['Umur'] > 68) & (data['Umur'] <= 85), 'Umur'] = 4
    data.loc[(data['Umur'] > 85) & (data['Umur'] <= 90), 'Umur'] = 5
```

```python
for data in full_data:
    obj_study = {'10th':0,'11th':1,
                 '12th':2,
                 '1st-4th':3,
                 '5th-6th':4,
                 '7th-8th':5,
                 '9th':6,
                 'D3':7,
                 'D4':8,
                 'Doktor':9,
                 'Master':10,
                 'Pendidikan Tinggi':11,
                 'Sarjana':12,
                 'SD':13,
                 'Sekolah Professional':14,
                 'SMA':15}
    data['Pendidikan'] = data['Pendidikan'].replace(obj_study)
```

```python
for data in full_data:
    obj_stat = {'Belum Pernah Menikah':0,
                'Berpisah':1,
                'Cerai':2,
                'Janda':3,
                'Menikah':4,
                'Menikah LDR':5}
    data['Status Perkawinan'] = data['Status Perkawinan'].replace(obj_stat)
```

```python
for data in full_data:
    obj_work = {'Asisten Rumah Tangga':0,
                'Ekesekutif Managerial':1,
                'Mesin Inspeksi':2,
                'Pembersih':3,
                'Pemuka Agama':4,
                'Penjaga':5,
                'Perbaikan Kerajinan':6,
                'Petani':7,
                'Sales':8,
                'Servis Lainnya':9,
                'Spesialis':10,
                'Supir':11,
                'Tech-support':12,
                'Tentara':13}
    data['Pekerjaan'] = data['Pekerjaan'].replace(obj_work)
```

```python
for data in full_data:
    obj_sex = {'Laki2':0,'Perempuan':1}
    data['Jenis Kelamin'] = data['Jenis Kelamin'].replace(obj_sex)
```

```python
for data in full_data:
    data.loc[data['Jam per Minggu'] <=20, 'Jam per Minggu'] = 0
    data.loc[(data['Jam per Minggu'] > 20) & (data['Jam per Minggu'] <= 40), 'Jam per Minggu'] = 1
    data.loc[(data['Jam per Minggu'] > 40) & (data['Jam per Minggu'] <= 60), 'Jam per Minggu'] = 2
    data.loc[(data['Jam per Minggu'] > 60) & (data['Jam per Minggu'] <= 80), 'Jam per Minggu'] = 3
    data.loc[(data['Jam per Minggu'] > 80) & (data['Jam per Minggu'] <= 100), 'Jam per Minggu'] = 4
```

```python
fee = {'<= 6 juta':0,'> 6 juta':1}
train['Gaji'] = train['Gaji'].replace(fee)
```
